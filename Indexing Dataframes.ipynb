# Indexing Dataframes-Normal and Multilevel

```python
# Index is essentially a row level label,
# 
# Methods use to create Indexes 
# 1.Indices can be inferred- create new series without an index here we get numeric values, 
# 2.Explicitly - Load the data from csv file and specified the header
# 3. set_index() function- takes a list of columns and promotes those columns to index. It is destructive process
```

```python
# Load the dataset
import pandas as pd
df = pd.read_csv('olympics.csv', index_col = 0, skiprows=1)
```

```python
# Setting up the column name
for col in df.columns:
    if col[:2] == '01':
        df.rename(columns={col:'Gold'+col[4:]}, inplace = True)
    if col[:2] == '02':
        df.rename(columns={col:'Silver'+col[4:]}, inplace = True)
    if col[:2] == '03':
        df.rename(columns={col:'Bronze'+col[4:]}, inplace = True)
    if col[:1] == '#':
        df.rename(columns={col:'#'+col[4:]}, inplace = True)
df.head()
```

```python
# set_index in play
# since set_index is destructive so we have to presive the original index
df['country'] = df.index
df = df.set_index('Gold')
df.head()
# New first row will added i.e Gold 
# Gold row has nothing
```

```python
# We can get rid of the index completely by calling the function reset_index.
# This promotes the index into a column and creates a default numbered index.
```

```python
df = df.reset_index()
df.head()
```

# MultiLevel Indexing

```python
# It is similar to composite keys in the relational Database
# To create a multi-level index, we simply call set index and give it a list of columns 
# that we're interested in promoting to an index. 
# Pandas will search through these in order, finding the distinct data and forming composite indices.
```

```python
# Change dataset for better understanding
df = pd.read_csv('census.csv')
df.head()
```

```python
# More about this file
#  this is a breakdown of the population level data at the US county level. 
#  in this data set there are two summarized levels, one that contains summary data for the whole country.
# And one that contains summary data for each state, and one that contains summary data for each county. 
```

```python
# see the unique sumlevel
df['SUMLEV'].unique()
# Two unique sumlevel 40 and 50
```

```python
df = df[df["SUMLEV"] == 50]
df.head()
```

```python
columns_to_keep = ['STNAME',
                  'CTYNAME',
                  'BIRTHS2010',
                  'BIRTHS2011',
                  'BIRTHS2012',
                  'BIRTHS2013',
                  'BIRTHS2014',
                  'BIRTHS2015',
                  'POPESTIMATE2010',
                  'POPESTIMATE2011',
                  'POPESTIMATE2012',
                  'POPESTIMATE2013',
                  'POPESTIMATE2014',
                  'POPESTIMATE2015']
df = df[columns_to_keep]
df.head()
```

```python
# Multi Level Index
df = df.set_index(['STNAME', 'CTYNAME'])
df.head()
```

```python
# Now how to Query to the multilevel index 
# First level is index 0 and second is index 1 and so on
df.loc['Alabama', 'Barbour County']
#df.loc['Michigan']
#df.loc['Michigan', 'Washtenaw County']
#df.loc['Alabama', 'Autauga County']
```

```python
# Since we have a MultiIndex of two values, the state and the county, 
# we need to provide two values as each element of our filtering list. 
# Note the big Bracket
df.loc[[('Alabama', 'Barbour County'), ('Alabama', 'Autauga County')]]
```

```python
# Some complex problem
# Multilevel index - create multilevel index with other name than existing column name and 
# appened or add some new value to new DataFrame
# Question- Reindex the purchase record DataFrame to be indexed hierarchically, fist by store, then by persion.
# Name these indexes 'Location' and 'Name'. Then add a new entry to it with the value of:
# Name: 'Kevyn', Item Purchased :'Kitty Food', Cost: 3.00, Location: 'Store 2'.
```

```python
purchase_1 = pd.Series({'Name': 'Chris',
                        'Item Purchased': 'Dog Food',
                        'Cost': 22.50})
purchase_2 = pd.Series({'Name': 'Kevyn',
                        'Item Purchased': 'Kitty Litter',
                        'Cost': 2.50})
purchase_3 = pd.Series({'Name': 'Vinod',
                        'Item Purchased': 'Bird Seed',
                        'Cost': 5.00})

df = pd.DataFrame([purchase_1, purchase_2, purchase_3], index=['Store 1', 'Store 1', 'Store 2'])


df = df.set_index([df.index, 'Name'])
df.index.names = ['Location', 'Name']
df = df.append(pd.Series(data={'Cost': 3.00, 'Item Purchased': 'Kitty Food'}, name=('Store 2', 'Kevyn')))
df
```
